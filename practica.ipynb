```python
import pandas as pd
```

```python
df = pd.read_csv('Bike Data.csv')
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

# Moment del dia
def get_moment_of_day(hour):
    if 6 <= hour < 12:
        return "Matí"
    elif 12 <= hour < 18:
        return "Tarda"
    elif 18 <= hour <= 23:
        return "Nit"
    else:
        return "Madrugada"

df["Moment del dia"] = df["Timestamp"].dt.hour.apply(get_moment_of_day)

# Dia de la setmana
df["Dia de la setmana"] = df["Timestamp"].dt.day_name(locale="ca_ES")  # En català

# Hora
df["Hora"] = df["Timestamp"].dt.hour

# Cap de setmana
df["Cap de setmana"] = df["Dia de la setmana"].isin(["Dissabte", "Diumenge"])

# Percentatge d'ocupació
df["Percentatge d'ocupació"] = (
    df["Bikes Available"] / (df["Bikes Available"] + df["Spaces Available"]) * 100
)

# Densitat d'estació
df["Densitat d'estació"] = df["Bikes Available"] + df["Spaces Available"]

# Classificació d'ús
def classify_usage(percentage):
    if percentage > 80:
        return "Alta ocupació"
    elif 50 <= percentage <= 80:
        return "Mitjana ocupació"
    else:
        return "Baixa ocupació"

df["Classificació d'ús"] = df["Percentatge d'ocupació"].apply(classify_usage)



# Resultat
display(df)
```

```python
df['Mida estacio'] = df['Bikes Available'] + df['Spaces Available']

def classify_size(mida):
    if mida > 40:
        return "Gran"
    elif 20 <= mida <= 40:
        return "Mitjana"
    else:
        return "Petita"

df["Classificació estacio"] = df["Mida estacio"].apply(classify_size)
```
